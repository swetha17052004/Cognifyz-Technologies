```python
import pandas as pd
import os
import matplotlib.pyplot as plt
full_file_path = r"C:\Users\Sweth\OneDrive\Desktop\DA inten\Dataset  (1).csv"
data = pd.read_csv(full_file_path)
price_range_counts = data['Price range'].value_counts()
price_range_percentage = (price_range_counts / len(data)) * 100
print("Percentage of restaurants in each price range:")
print(price_range_percentage)
plt.figure(figsize=(10, 6))
price_range_counts.plot(kind='bar', color='lightcoral')
plt.title('Distribution of Restaurants by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
output_file_path = os.path.join(r"C:\Users\Sweth\OneDrive\Desktop\DA inten", 'Price_Range_Distribution.png')
plt.savefig(output_file_path)
plt.show()
price_range_percentage.to_csv(os.path.join(r"C:\Users\Sweth\OneDrive\Desktop\DA inten", 'Price_Range_Percentage.csv'))
print(f"The price range distribution bar chart has been saved to {output_file_path}.")
print(f"The price range percentage data has been saved to 'Price_Range_Percentage.csv'.")
```
