```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
full_file_path = r"C:\Users\Sweth\OneDrive\Desktop\DA inten\Dataset  (1).csv"
data = pd.read_csv(full_file_path)
try:
    online_delivery_counts = data['Has Online delivery'].value_counts()
    total_restaurants = len(data)
    online_delivery_percentage = (online_delivery_counts.get('Yes', 0) / total_restaurants) * 100
    print(f"Percentage of restaurants offering online delivery: {online_delivery_percentage:.2f}%")
    average_rating_with_delivery = data[data['Has Online delivery'] == 'Yes']['Aggregate rating'].mean()
    average_rating_without_delivery = data[data['Has Online delivery'] == 'No']['Aggregate rating'].mean()
    print(f"Average rating for restaurants with online delivery: {average_rating_with_delivery:.2f}")
    print(f"Average rating for restaurants without online delivery: {average_rating_without_delivery:.2f}")
    sns.set(style="whitegrid")
    labels = ['Offers Online Delivery', 'No Online Delivery']
    sizes = [online_delivery_counts.get('Yes', 0), online_delivery_counts.get('No', 0)]
    colors = ['#66b3ff', '#99ff99']
    plt.figure(figsize=(7, 7))
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors, wedgeprops={'edgecolor': 'black'})
    plt.title("Percentage of Restaurants Offering Online Delivery")
    plt.show()
    ratings_df = pd.DataFrame({
        'Delivery Option': ['With Online Delivery', 'Without Online Delivery'],
        'Average Rating': [average_rating_with_delivery, average_rating_without_delivery]
    })

    plt.figure(figsize=(8, 6))
    sns.barplot(x='Delivery Option', y='Average Rating',data=ratings_df)
    plt.title("Average Rating Comparison: Online Delivery vs No Online Delivery")
    plt.ylabel("Average Rating")
    plt.ylim(0, 5)  # Assuming the rating scale is from 0 to 5
    plt.show()

except KeyError as e:
    print(f"Error: The dataset does not contain expected columns. Missing column: {e}")
except Exception as e:
    print(f"An error occurred: {e}")
```
