```python
import pandas as pd
import os
import matplotlib.pyplot as plt

# Define the file path
full_file_path = r"C:\Users\Sweth\OneDrive\Desktop\DA inten\Dataset  (1).csv"

# Load the dataset
data = pd.read_csv(full_file_path)

# 1. Identify the city with the highest number of restaurants
city_counts = data['City'].value_counts()
city_with_most_restaurants = city_counts.idxmax()
num_restaurants = city_counts.max()

# 2. Calculate the average rating for restaurants in each city
average_ratings = data.groupby('City')['Aggregate rating'].mean()

# 3. Determine the city with the highest average rating
city_with_highest_avg_rating = average_ratings.idxmax()
highest_avg_rating = average_ratings.max()

# Print the results
print(f"The city with the highest number of restaurants is {city_with_most_restaurants} with {num_restaurants} restaurants.")
print("Average rating for each city:")
print(average_ratings)
print(f"The city with the highest average rating is {city_with_highest_avg_rating} with an average rating of {highest_avg_rating:.2f}.")

# Save the results to a new CSV file
root = r"C:\Users\Sweth\OneDrive\Desktop\DA inten"
output_file_path = os.path.join(root, "City_Analysis_Results.csv")
results = {
    "City with Most Restaurants": [city_with_most_restaurants],
    "Number of Restaurants": [num_restaurants],
    "City with Highest Average Rating": [city_with_highest_avg_rating],
    "Highest Average Rating": [highest_avg_rating]
}

# Convert the dictionary to a DataFrame
results_df = pd.DataFrame(results)

# Save the summary results and average ratings to separate CSV files
results_df.to_csv(output_file_path, index=False)
average_ratings.to_csv(os.path.join(root, "City_Average_Ratings.csv"))

# Plot top 10 cities by number of restaurants
top_10_restaurants = city_counts.head(10)
plt.figure(figsize=(10, 6))
top_10_restaurants.plot(kind='bar', color='skyblue')
plt.title('Top 10 Cities by Number of Restaurants')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig(os.path.join(root, 'Top_10_Cities_by_Restaurants.png'))
plt.show()

# Plot top 10 cities by average rating
top_10_ratings = average_ratings.sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
top_10_ratings.plot(kind='bar', color='lightgreen')
plt.title('Top 10 Cities by Average Rating')
plt.xlabel('City')
plt.ylabel('Average Rating')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig(os.path.join(root, 'Top_10_Cities_by_Avg_Rating.png'))
plt.show()

print(f"Results have been saved to {output_file_path} and City_Average_Ratings.csv.")
```
